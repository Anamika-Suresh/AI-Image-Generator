```python

import os
import datetime
import utils

USERNAME = "ANAMIKA"
PASSWORD = "1234"

user = input("Enter username: ")
pwd = input("Enter password: ")

if user == USERNAME and pwd == PASSWORD:
    print("Login successful!")

    def get_int_input(prompt):
      while True:
         try:
             return int(input(prompt))
         except ValueError:
             print("Invalid input! Please enter an integer.")

    def get_float_input(prompt):
       while True:
         try:
            return float(input(prompt))
         except ValueError:
            print("Invalid input! Please enter a number.")
    class Student:

       def __init__(self, roll_no, name, age, department, marks, time_of_entry=None):
         self.roll_no = roll_no
         self.name = name
         self.age = age
         self.department = department
         self.marks = marks
         self.time_of_entry = time_of_entry or datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
       def display(self):
         print(f"Roll No: {self.roll_no}")
         print(f"Name   : {self.name}")
         print(f"Age    : {self.age}")
         print(f"Dept   : {self.department}")
         print(f"Marks  : {self.marks}")
         print(f"Time of Entry  : {self.time_of_entry}")
         print("-" * 30)
    class StudentManagementSystem:
       def __init__(self,filename="students.txt"):
         self.students = []
         self.filename = filename
         self.next_id = 1
         self.load_students()

       def save_students(self):
        """Save all student records to a text file"""
        with open(self.filename, "w") as file:
          for student in self.students:
             file.write(f"{student.roll_no},{student.name},{student.age},{student.department},{student.marks},{student.time_of_entry}\n")

       def load_students(self):
        "Load student records from text file"
        if os.path.exists(self.filename):
          with open(self.filename, "r") as file:
             for line in file:
                roll_no, name, age, department, marks,time_of_entry = line.strip().split(",")
                student = Student(roll_no, name, int(age), department, float(marks), time_of_entry)
                self.students.append(student)
             if self.students:
                    last_id_num = int(self.students[-1].roll_no[3:])
                    self.next_id = self.students[-1].roll_no + 1
       def add_student(self):
        """Add a new student record"""
        try:
          roll_no = int(input("Enter Roll Number: "))
          name = input("Enter Name: ")
          age = int(input("Enter Age: "))
          department = input("Enter Department: ")
          marks = float(input("Enter Marks: "))

          student = Student(roll_no, name, age, department, marks)
          self.students.append(student)
          self.save_students()
          print(f"Student record added successfully!(Auto-Generated Student ID: {roll_no})\n")
        except ValueError:
          print("Invalid input! Please enter numeric values for Roll Number, Age, and Marks.\n")

       def view_student(self):
          """View all student records"""
          if not self.students:
            print(" No student records found!\n")
          else:
            print("--- Student Records ---")
            for student in self.students:
                student.display()

       def update_student(self):
        try:
           roll_no = int(input("Enter Student Roll Number to update: "))
           for student in self.students:
             if student.roll_no == roll_no:
                print("Student found! Enter new details:")

                new_name = input(f"New Name ({student.name}): ") or student.name
                new_age = input(f"New Age ({student.age}): ") or student.age
                new_department = input(f"New Department ({student.department}): ") or student.department
                new_marks = input(f"New Marks ({student.marks}): ") or student.marks

                student.name = new_name
                student.age = int(new_age)
                student.department = new_department
                student.marks = float(new_marks)

                self.save_students()
                print(" Student record updated successfully!\n")
                return
           print("Student with given Roll Number not found!\n")
        except ValueError:
         print(" Invalid input! Please enter numeric values for Roll Number, Age, and Marks.\n")

       def delete_student(self):
         "Delete a student based on Roll Number"
         try:
            roll_no = int(input("Enter Roll Number of student to delete: "))
            for student in self.students:
               if student.roll_no == roll_no:
                  self.students.remove(student)
                  self.save_students()
                  print(f" Student with Roll Number {roll_no} deleted successfully!\n")
                  return
            print("Student with given Roll Number not found!\n")
         except ValueError:
            print("Invalid input! Please enter a numeric Roll Number.\n")

       def search_student(self):
        "Search student by name or department"
        keyword = input("Enter Name or Department to search: ").lower()
        found = False
        for student in self.students:
            if keyword in student.name.lower() or keyword in student.department.lower():
                student.display()
                found = True
        if not found:
            print(" No matching student found!\n")

       def top_students(self):

        "View top-performing students by marks"
        if not self.students:
            print(" No student records found!\n")
            return

        sorted_students = sorted(self.students, key=lambda s: s.marks, reverse=True)
        print("Top Performing Students:")
        for student in sorted_students[:3]:
            student.display()
    sms = StudentManagementSystem()
    while True:
     print(" Student Management System ")
     print("1. Add New Student")
     print("2. View Students")
     print("3. Update Student")
     print("4. Delete Student")
     print("5. Search Student")
     print("6. View Top Students")
     print("7. Exit")
     choice = input("Enter your choice: ")

     if choice == "1":
        sms.add_student()
     elif choice == "2":
        sms.view_student()
     elif choice == "3":
        sms.update_student()
     elif choice == "4":
        sms.delete_student()
     elif choice == "5":
        sms.search_student()
     elif choice == "6":
        sms.top_students()
     elif choice == "7":
        print("Thank you for using the Student Management System")
        break
     else:
        print("Invalid choice! Please try again.")

    with open("students.txt") as my_file:
     print("File Content")
     print(my_file.read())
else:
    print("Invalid username or password.")
    exit()
```
